```python
import os
import pandas as pd
from sodapy import Socrata

# 1. Handle Secrets for both Colab and GitHub
try:
    # This only runs inside Google Colab
    from google.colab import userdata
    cambridge_city = userdata.get('cambridge_city')
    print("Environment: Google Colab")
except (ImportError, ModuleNotFoundError):
    # This only runs inside GitHub Actions
    cambridge_city = os.getenv('CAMBRIDGE_CITY_SECRET')
    print("Environment: GitHub Actions")

# 2. Check for missing keys
if not cambridge_city:
    raise ValueError("API Key not found! Check your Colab Secrets or GitHub Secrets.")

# 3. Initialize the Client
client = Socrata("data.cambridgema.gov", cambridge_city)
```

```python
# Define the query with the specific filters requested
# Dataset ID: 2z9k-mv9g
query = """
SELECT
    *
WHERE
    issue_type = 'Rodent Sighting'
ORDER BY
    ticket_created_date_time DESC
LIMIT
    300
"""

# Fetch the data
results = client.get("2z9k-mv9g", query=query)

# Convert to DataFrame
results_df = pd.DataFrame.from_records(results)

# Display the results
display(results_df)

# Verify the issue types
if not results_df.empty:
    print("\nCounts by Issue Type:")
    print(results_df['issue_type'].value_counts())
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Calculate sentiment scores (using the compound score)
# We convert to string to handle any potential missing values gracefully
results_df['sentiment_score'] = results_df['issue_description'].apply(lambda x: sid.polarity_scores(str(x))['compound'])

# Display the results
display(results_df[['issue_description', 'sentiment_score']].head())
```

```python
# Clean up the html_url column
# This logic checks if the value is a dictionary (raw data) and extracts the 'url'.
# If it's already a string (e.g. if the cell is re-run), it leaves it as is.
results_df['html_url'] = results_df['html_url'].apply(lambda x: x.get('url') if isinstance(x, dict) else x)

# Display the first few rows of the modified column
display(results_df[['html_url']].head())
```

```python
import numpy as np

# Ensure datetime columns are in correct format
results_df['ticket_created_date_time'] = pd.to_datetime(results_df['ticket_created_date_time'])
results_df['ticket_closed_date_time'] = pd.to_datetime(results_df['ticket_closed_date_time'])

# Define tomorrow's date
tomorrow = pd.Timestamp.now() + pd.Timedelta(days=1)

# Logic:
# If status is Closed or Archived -> Closed Date - Created Date
# Else -> Tomorrow - Created Date

# Create a boolean mask for closed/archived tickets
is_closed = results_df['ticket_status'].isin(['Closed', 'Archived'])

# Calculate time deltas
# We use pd.Timedelta(days=1) to convert the timedelta to a float representation of days
results_df['response_time'] = np.where(
    is_closed,
    (results_df['ticket_closed_date_time'] - results_df['ticket_created_date_time']) / pd.Timedelta(days=1),
    (tomorrow - results_df['ticket_created_date_time']) / pd.Timedelta(days=1)
)

# Display the results to verify
display(results_df[['ticket_status', 'ticket_created_date_time', 'ticket_closed_date_time', 'response_time']].head())
```
